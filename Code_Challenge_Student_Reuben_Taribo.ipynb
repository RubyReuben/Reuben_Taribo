#### Q6. What are the top 5 lowest CO2 Emission Grams/Mile emmisions of cars for each of the following brands: Toyota, Ford, Volkswagen, Nissan, Honda
<b>Format:</b> A 5X6 list with the first element of each row being the Make of the cars and the following five values being floats sorted in ascending order. The Makes should appear in order listed in the question starting with Ford and ending with Volkswagen (see example).

Example answer:
>[['Toyota', 100.0, 140.0, 140.0, 150.0, 150.0],  
 ['Ford',
  100.025641025641,
  200.677633075759575,
  200.677633075759575,
  200.677633075759575,
  200.677633075759575],  
 ['Volkswagen', 139.0, 154.0, 166.5, 166.5, 166.5],  
 ['Nissan', 122.0, 122.0, 122.0, 122.0, 160.0],  
 ['Honda', 100.0, 100.0, 100.0, 100.0, 123.91684618442447]]

```python
########## Q6
#Your code here
cartype = ['Toyota', 'Ford', 'Volkswagen', 'Nissan', 'Honda']

carlen = 0

for item in cartype:
  carlen = carlen + 1

result = []
for i in cartype:
  r = df.query('Make == @i')['CO2 Emission Grams/Mile'].reset_index().sort_values(["CO2 Emission Grams/Mile"], ascending=[True]).head(5)
  r.drop("index",axis=1,inplace=True)
  result.append(r)
result = np.array(result).tolist()

flat_list = []
for sublist in result:
    for item in sublist:
      for j in item:
          flat_list.append(j)

counter = 0
count = 0
while count < carlen:
    flat_list.insert(counter,cartype[count])
    counter = counter + (carlen+1)
    count = count + 1

answer_dict['Q6'] = np.array(flat_list, dtype="O").reshape((5, 6)).tolist()
 
answer_dict['Q6']
```

#### Q7. Form 7 groups of 5 years to calculated the median Combined MPG of each group. The first group is from 1984 to 1988, the second from 1989 to 1993 and so on. The last group will have years not appearing in the dataset.
Note: The groups ranges are inclusive on both sides, the first group starts with 1984 and cars from 1984 are included in it.

<b>Format:</b> A 7X2 list with the first element of each row being a tuple of two integers being the lower and uppper range of the year groups and the esecond element being the median Combined MPG of that group, a float number.

Example answer:
>[[(1984, 1988), 11.0],  
 [(1989, 1993), 10.0],  
 [(1994, 1998), 10.0],  
 [(1999, 2003), 14.0],  
 [(2004, 2008), 13.0],  
 [(2009, 2013), 14.0],  
 [(2014, 2018), 15.0]]

```python
########## Q7
#Your code here
r = np.sort(df['Year'].unique()).tolist()

startyear = r[0]
endyear = startyear + 4

lst = []

for i in r:
  if startyear < r[-1]:
    mask = df.query('Year >= @startyear & Year <= @endyear')['Combined MPG'].median()
    y = [startyear,endyear]
    y = tuple(y)
    lst.append(y)
    lst.append(mask)
    startyear = endyear + 1
    endyear = startyear + 4
answer_dict['Q7'] = np.array(np.reshape(lst, (7,(2)))).tolist()
answer_dict['Q7']
```
